```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading ipl matches data
ipl = pd.read_csv('matches.csv')
ipl.head()
```

```python
#checking shape of dataframe
ipl.shape
```

```python
#getting frequency of most man of match winner
ipl['player_of_match'].value_counts()[:5]
```

```python
#barplot man of match frequency
plt.figure(figsize=(5,5))
sns.barplot(x=ipl['player_of_match'].value_counts()[:5].keys(), y=ipl['player_of_match'].value_counts()[:5])
plt.title('Top 5 Man of Match Winner')
plt.xticks(rotation=45)
plt.show()
```

```python
ipl['result'].value_counts()
```

```python
ipl['toss_winner'].value_counts()
ipl['winner'].value_counts()
ipl[ipl['toss_winner']== ipl['winner']].shape
```

```python
batting_first = ipl[ipl['win_by_runs']!= 0]
```

```python
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])
plt.show()
```

```python
#batting first and winning match
batting_first['winner'].value_counts()[:5]

sns.barplot(x=batting_first['winner'].value_counts()[:5].keys(), y=batting_first['winner'].value_counts()[:5])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(x=batting_first['winner'].value_counts(), labels=batting_first['winner'].value_counts().keys(), autopct='%0.1f%%')
```

```python
batting_second = ipl[ipl['win_by_wickets']!= 0]
batting_second.shape
```

```python
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'], bins=30)
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
plt.figure(figsize=(7,7))
sns.barplot(x=batting_second['winner'].value_counts()[0:5].keys(),
y=batting_second['winner'].value_counts()[:5])
plt.title('Winning after batting Second')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(x=batting_second['winner'].value_counts(), 
labels=batting_second['winner'].value_counts().keys(),autopct='%0.1f%%')
plt.show()
```
