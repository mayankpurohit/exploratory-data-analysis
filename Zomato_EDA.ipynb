```python
zomato_country = pd.read_excel('Country-Code.xlsx')
zomato_country
```

```python
zomato = pd.merge(zomato,zomato_country,on='Country Code', how='left')
```

```python
zomato.head()
```

```python
zomato.columns
```

```python
country_names = zomato['Country'].value_counts().index
country_values = zomato['Country'].value_counts().values

```

```python
sns.barplot(x=country_names[:10], y=country_values[:10])
plt.xlabel('Country')
plt.xticks(rotation=90)
```

```python
ratings = zomato.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count', data=ratings)
plt.xticks(rotation=90)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color',data=ratings)
plt.xticks(rotation=90)
```

```python
sns.countplot(x='Rating color', data=ratings,palette=['blue','red','orange','yellow','green', 'green'])
```
