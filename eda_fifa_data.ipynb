```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa_df = pd.read_csv('players_20.csv')
fifa_df.head()
```

```python
fifa_df.shape
```

```python
#top 10 countries in fifa20
fifa_df['nationality'].value_counts()[:10]
```

```python
plt.figure(figsize=(7,7))
# plt.bar(x=list(fifa_df['nationality'].value_counts()[:10].keys()),y=fifa_df['nationality'].value_counts()[:10].values)
sns.barplot(x=fifa_df['nationality'].value_counts()[:10].keys(),y=fifa_df['nationality'].value_counts()[:10].values)
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_earner = fifa_df[['short_name','wage_eur']].sort_values(by=['wage_eur'],ascending=False)[:10]
top_10_earner
```

```python
sns.barplot(x=top_10_earner['short_name'], y=top_10_earner['wage_eur'])
plt.xticks(rotation=45)
plt.show()
```

```python
england = fifa_df[fifa_df['nationality'] == 'England']
england.head(10)
```

```python
england.sort_values(by=['height_cm'], ascending=False).head()
```

```python
england.sort_values(by=['weight_kg'], ascending=False).head()
```

```python
top_earner_england = england[['short_name', 'wage_eur']].sort_values(by=['wage_eur'], ascending=False)[:10]
```

```python
sns.barplot(x=top_earner_england['short_name'], y=top_earner_england['wage_eur'])
plt.xticks(rotation=45)
plt.show()
```
